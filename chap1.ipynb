```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import mglearn
from pandas.plotting import scatter_matrix

iris_dataset = load_iris()

print(iris_dataset.keys()) #dict_keys(['data', 'target', 'frame', 'target_names', 'DESCR', 'feature_names', 'filename', 'data_module'])
print(iris_dataset['target_names']) #予測しようとしている花の種類：['setosa' 'versicolor' 'virginica']→0,1,2のラベルに対応してる
print(iris_dataset['feature_names'])#特徴量の名前：がく（sepal）の長さ，がくの幅,花弁(petal)の長さ，花弁の幅
# でーた本体はdata(特徴量4つ)x150,target(それぞれのラベルとして0,1,2)x150に格納されている


from sklearn.model_selection import train_test_split # データを自動的に訓練用と検証用に分けるためのライブラリ：割合は75：25らしい
X_train,X_test, y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0) #大文字X：データ，小文字y：ラベル, random_state:同じ乱数シードによってランダムに取り出している

import pandas as pd
# 元データの内訓練データのみデータフレームにする
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)

# pandasのscatter matrix：
# grr = pd.scatter_matrix(iris_dataframe,c=y_train) #c:color
fig, ax = plt.subplots()
grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(10, 10),
                        ax=ax, marker='o', hist_kwds ={'bins': 20}, s = 60,
                        alpha=0.8, cmap=mglearn.cm3)
plt.show()
```

```python
from sklearn import neighbors
# k-最近傍法によるクラス分類
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```
