```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import mglearn
from pandas.plotting import scatter_matrix

iris_dataset = load_iris()

print(iris_dataset.keys()) #dict_keys(['data', 'target', 'frame', 'target_names', 'DESCR', 'feature_names', 'filename', 'data_module'])
print(iris_dataset['target_names']) #['setosa' 'versicolor' 'virginica']0,1,2
print(iris_dataset['feature_names'])#sepal,(petal)
# data(4)x150,target(0,1,2)x150


from sklearn.model_selection import train_test_split # 7525
X_train,X_test, y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0) #Xy, random_state:

import pandas as pd
# 
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)

# pandasscatter matrix
# grr = pd.scatter_matrix(iris_dataframe,c=y_train) #c:color
fig, ax = plt.subplots()
grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(10, 10),
                        ax=ax, marker='o', hist_kwds ={'bins': 20}, s = 60,
                        alpha=0.8, cmap=mglearn.cm3)
plt.show()
```
